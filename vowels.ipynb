```python
name = input("Enter your name: ")

vowels = "aeiouAEIOU"
letter_count = {}
vowel_count = 0

for ch in name:
    # Count letters using dictionary
    if ch in letter_count:
        letter_count[ch] += 1
    else:
        letter_count[ch] = 1

    # Check vowel
    if ch in vowels:
        vowel_count += 1

print("\nVowels in the name:")
for ch in name:
    if ch in vowels:
        print(ch)

print("\nTotal number of vowels:", vowel_count)

print("\nCount of each letter:")
for key, value in letter_count.items():
    print(key, ":", value)
```
